```python
from sklearn.preprocessing import normalize
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
#read in dataset
df = pd.read_csv('myopia.csv')

#printing dataset
df
```

```python
#drop myopic column
df2 = df.drop('MYOPIC',axis = 1)

#print dataset
df2
```

```python
#standardize dataset
norm = normalize(df2)
```

```python
norm
```

```python
#create pca model
pca = PCA(n_components=0.90)

pcamyopic = pca.fit_transform(norm)

#turn pca into a dataframe
dfpca = pd.DataFrame(
    data=pcamyopic,
    columns=['principal component1', 'principal component2', 'principal component3'],
)
dfpca.head()
```

```python
#run t-SNE
tsne = TSNE(learning_rate=300)
tsne_feat = tsne.fit_transform(pcamyopic)
```

```python
# Check dimensions
tsne_feat.shape
```

```python
#create scatter plot of the output
plt.scatter(tsne_feat[:,0], tsne_feat[:,1])
plt.show()
```

```python
#convert tsne into a dataframe
tsne_df = pd.DataFrame(
    data=tsne_feat,
    columns=['principal component 1', 'principal component 2'],
)
tsne_df.head()
```

```python
#Create a for loop to determine the inertia for each k between 1 through 10
inertia = []
k = list(range(1, 11))

#find the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=50)
    km.fit(tsne_df)
    inertia.append(km.inertia_)

#Create the Elbow Curve data
curve_data = {'k': k, 'inertia': inertia}
curve_df = pd.DataFrame(curve_data)
```

```python
#plot the elbow curve to find the best candidate s for k
plt.plot(curve_df['k'], curve_df['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve for Myopic Data')
plt.show()
```

```python
def get_clusters(k, data):
    #initialize the k Means model
    model = KMeans(n_clusters=k, random_state=42)

    #train the model
    model.fit(data)

    #predict clusters
    predict = model.predict(data)

    #create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
clusters = get_clusters(4, tsne_df)
```
